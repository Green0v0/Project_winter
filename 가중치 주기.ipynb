```python
# 노인복지시설수 가중치
result_before['복지시설가중치']=''

for i in list5:
    if result_before['복지시설개수'][i].item()==0:
        result_before['복지시설가중치'][i]=7
    elif result_before['복지시설개수'][i].item()==1:
        result_before['복지시설가중치'][i]=6
    elif result_before['복지시설개수'][i].item()==2:
        result_before['복지시설가중치'][i]=5
    elif result_before['복지시설개수'][i].item()==3:
        result_before['복지시설가중치'][i]=4
    elif result_before['복지시설개수'][i].item()==4:
        result_before['복지시설가중치'][i]=3
    elif result_before['복지시설개수'][i].item()==5:
        result_before['복지시설가중치'][i]=2
    elif result_before['복지시설개수'][i].item()==6:
        result_before['복지시설가중치'][i]=1
    else:
        result_before['복지시설가중치'][i]=0
        
result_before
```

```python
# 필요주소 붙이기
count = 0
for j in range(0, len(urban)):
    for i in range(0, len(result_before)):
        if result_before.iloc[i, 0] in urban.iloc[j, 0]:
            count = count+1
            urban.iloc[j, 3]=result_before.iloc[i, 0]

print(count)
```
